```python
from collections import deque
import networkx as nx
import matplotlib.pyplot as plt

graph = {
    "Islamabad": ["Rawalpindi", "Lahore", "Peshawar"],
    "Rawalpindi": ["Islamabad", "Peshawar", "Quetta"],
    "Peshawar": ["Islamabad", "Rawalpindi", "Quetta"],
    "Lahore": ["Islamabad", "Multan", "Quetta"],
    "Multan": ["Lahore", "Karachi", "Quetta"],
    "Quetta": ["Rawalpindi", "Peshawar", "Multan", "Karachi"],
    "Karachi": ["Multan", "Quetta"]
}


def bfs_shortest_path(graph, start, goal):

    queue = deque([(start, [start])])
    while queue:
        (city, path) = queue.popleft()


        for neighbor in graph[city]:
            if neighbor in path:
                continue
            if neighbor == goal:
                return path + [neighbor]
            else:
                queue.append((neighbor, path + [neighbor]))

    return None

def draw_graph(graph, path=None):
    G = nx.Graph()


    for city in graph:
        for neighbor in graph[city]:
            G.add_edge(city, neighbor)


    pos = {
        "Islamabad": (1, 4),
        "Rawalpindi": (2, 4),
        "Peshawar": (1, 5),
        "Lahore": (1, 3),
        "Multan": (1, 2),
        "Quetta": (2, 3),
        "Karachi": (1, 1)
    }


    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=3000, font_size=10, font_weight='bold', edge_color="gray")


    if path:
        path_edges = [(path[i], path[i + 1]) for i in range(len(path) - 1)]
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="blue", width=3)

    plt.title("City Connections")
    plt.show()


start_city = "Islamabad"
goal_city = "Karachi"
shortest_path = bfs_shortest_path(graph, start_city, goal_city)


if shortest_path:
    print("Shortest path from {} to {}: {}".format(start_city, goal_city, " -> ".join(shortest_path)))

    draw_graph(graph, shortest_path)
else:
    print("No path found from {} to {}".format(start_city, goal_city))
    draw_graph(graph)

```
